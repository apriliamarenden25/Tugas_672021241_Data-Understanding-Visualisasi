**SOAL 2 : PERMEN**

```python
import matplotlib.pyplot as plt

nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
jumlah_permen = [52, 39, 78, 13, 78]

total_permen = sum(jumlah_permen)
peluang_kopiko = jumlah_permen[nama_permen.index('Kopiko')] / total_permen * 100


labels = nama_permen
sizes = jumlah_permen
colors = ['#A5DD9B','#1B1A55','#F6F193','#FBA834','#9195F6']
explode = (0, 0.1, 0, 0, 0)
plt.figure(figsize=(15, 5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Peluang Memilih Permen Kopiko')
plt.show()
```

**SOAL 3: MAKANAN**

```python
import matplotlib.pyplot as plt

datapenjualan_makananpenutup = {
    'Donat': 14,
    'Pastel': 5,
    'Kue Coklat': 12,
    'Ice Cream': 19,
    'Puding Vanila': 6,
    'Brownies': 8,
    'Puding Strawberi': 12,
    'Puding Coklat': 9,
    'Ice Cream Nutela': 10,
    'Kue Coklat-Keju': 17,
    'Kue Wajik': 2,
    'Kue Sus': 9,
    'Mochi': 13,
}

sorted_datapenjualan = sorted(datapenjualan_makananpenutup.items(), key=lambda x: x[1], reverse=True)
hapus_item = sorted_datapenjualan[-3:]


for item in hapus_item:
    del datapenjualan_makananpenutup[item[0]]


plt.figure(figsize=(15, 5))
plt.bar(datapenjualan_makananpenutup.keys(), datapenjualan_makananpenutup.values(), color='skyblue')
plt.title('Popularitas Makanan Penutup')
plt.xlabel('Makanan Penutup')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()

print("Makanan penutup yang harus disingkirkan:")
for item in hapus_item:
    print("- ", item[0])
```
